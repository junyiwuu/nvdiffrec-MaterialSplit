```python
import torch
from pytorch_msssim import ms_ssim, ssim
from PIL import Image
from torchvision import transforms
import matplotlib.pyplot as plt
```

```python
number = "000020"
pass_name = "mesh"
# img = Image.open(f"../output/spot_1/relmse-25_06_17-21_45/evaluation/mesh_validate/val_{number}_opt.png")
# target = Image.open(f"../output/spot_1/relmse-25_06_17-21_45/evaluation/mesh_validate/val_{number}_ref.png")

img = Image.open(f"../output/spot_1/relmse-25_06_17-21_45/evaluation/{pass_name}_validate/val_{number}_opt.png")
target = Image.open(f"../output/spot_1/relmse-25_06_17-21_45/evaluation/{pass_name}_validate/val_{number}_ref.png")

fig, axes = plt.subplots(1,2,figsize=(10,5))
axes[0].imshow(img)
axes[1].imshow(target)
axes[0].axis('off')
axes[1].axis('off')

toTensor = transforms.ToTensor()
img = toTensor(img).unsqueeze(0)
target=toTensor(target).unsqueeze(0)
loss = ms_ssim(img, target, win_size=15)
print(f"msssim loss is {1000*(1-loss):.9f}")
print(f"mse loss is {1000*(torch.nn.MSELoss()(img, target)):.9f}")
```
